##### （1）编写函数 stat， 要求该函数同时计算均值、最大值、最小值、标准差、峰度和偏 度。 （提示： R 默认不提供函数计算峰度和偏度，可以自编公式或者使用 fBasics 包。可使用 skewness(x）计算 x 的偏度， kurtosis(x）计算 x 的峰度。）

```r
install.packages('fBasics')
library(fBasics)
stat <- function(t)
{
    avg_t <- mean(t)
    max_t = max(t)
    min_t = min(t)
    sd_t = sd(t)
    kurtosis_t = kurtosis(t)
    skewness_t = skewness(t)
    return(c(avg_t,max_t,min_t,sd_t,kurtosis_t,skewness_t))
}
stat(c (2, 6,4, 9, 12))    
```
