#### （1）数据预处理，调整数据类型，将年龄、工龄分组

```r
data <- read.csv("bankloan.csv", sep = ",", encoding = "UTF-8")
# data
```

```r
age <- cut(data$age,breaks = c(0, 20, 30, 40, 50, 60),include.lowest = TRUE)
age
workAge <- cut(data$seniority,breaks = c(0, 1, 3, 5, 10, 15, 20, 30, 40),include.lowest = TRUE)
workAge
study <- factor(data$education)
study
```

##### 1、绘制违约与不违约客户的年龄、教育和工龄的条形图

```r
# install.packages('RColorBrewer')
library(RColorBrewer)
```

```r
debt <- data$debt_rate / 100 * data$income
# debt
```

```r
norm_debt <- rnorm(1000, mean(debt), sd(debt))
hist(debt, freq = FALSE, breaks = 50)
lines(density(debt), col = 2)
lines(density(norm_debt), lty = 2, col = 3) 
legend("topright", c("density", "normal"), lty = 1:2, col = 2:3, bty = "n")
```

#### （4）绘制不同年龄、教育和工龄下客户的收入与负债的散点图

##### 7、绘制不同年龄、教育和工龄下客户的收入与负债的散点图

###### 7.1 绘制不同年龄下客户的收入与负债的散点图

```r
l_age <- levels(age)
op <- par(mfrow = c(1, 4))
for (i in 1:nlevels(age)) {
  plot(data$income[age == l_age[i]], debt[age == l_age[i]], 
       main = paste("age = ", l_age[i]), xlab = "income", ylab = "debt")
  abline(lm(debt[age == l_age[i]] ~ data$income[age == l_age[i]]), col = "red")
}
par(op)
```

###### 7.2 绘制不同教育下客户的收入与负债的散点图

```r
l_study <- levels(study)
op <- par(mfrow = c(1, 5))
for (i in 1:nlevels(study)) {
  plot(data$income[study == l_study[i]], debt[study == l_study[i]], 
       main = paste("study = ", l_study[i]), xlab = "income", ylab = "debt")
  abline(lm(debt[study == l_study[i]] ~ data$income[study == l_study[i]]), col = "red")
}
par(op)
```

###### 7.3 绘制不同客户下客户的收入与负债的散点图

```r
l_workAge <- levels(workAge)
op <- par(mfrow = c(1, 4))
for(i in 1:nlevels(workAge)){
  plot(data$income[workAge == l_workAge[i]], debt[workAge == l_workAge[i]], 
       main = paste("workAge = ", l_workAge[i]), xlab = "income", ylab = "debt")
  abline(lm(debt[workAge == l_workAge[i]] ~ data$income[workAge == l_workAge[i]]), col = "red")
}
par(op)
```
